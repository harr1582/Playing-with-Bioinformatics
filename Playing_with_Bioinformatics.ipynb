```
# 1.3
text = list_
print(text)
comp_nucs = []
for each in text:
  if each == 'A':
    comp_nucs.append('T')
  if each == 'T':
    comp_nucs.append('A')
  if each == 'C':
    comp_nucs.append('G')
  if each == 'G':
    comp_nucs.append('C')
print (comp_nucs)
merged = ''.join(comp_nucs)

def reverse(string_): 
    string_ = string_[::-1] 
    return string_
reverse(merged)
```
